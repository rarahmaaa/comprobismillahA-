**format specier**

```python
template = 'halo, saya %s dari %s'
print (template % ('rahma', 'indo'))
```

**fungsi format**

```python
template = 'halo, saya {nama} dari {asal}'
template_2 = 'saya suka makan {} dan minum {}'
print (template.format(nama = 'rahma', asal = 'indonesia'))
print (template_2.format('nasi goreng', 'teh pucuk'))
```
